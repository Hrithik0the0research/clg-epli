```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("data.csv")
data=data.dropna()
data
```

```python
colm=data.columns
len(colm)
```

```python
features_col=colm[1:179]
features_col
```

```python
class_col=colm[179]
class_col
```

```python
class_val=data[class_col]
```

```python
features=data[features_col]
features=features.dropna()
features
```

```python

x=features
```

```python
x=np.array(x)
```

```python

""""
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
x=st.fit_transform(features)
print(x)
"""
```

```python
x=x.reshape(x.shape[0],x.shape[1],1)
```

```python
y=class_val
print(y)
```

```python
y=list(y)
y
```

```python
y_val=[]
for i in y:
    if i!=1:
        y_val.append(0)
    else:
        y_val.append(1)
```

```python
y_val=np.array(y_val)
y_val.shape
```

```python
"""
from tensorflow.keras.utils import to_categorical
y=to_categorical(class_val)
y
"""
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y_val,test_size=0.3,random_state=42)
```

```python
print(x_train.shape,y_train.shape)
```

```python
from keras.layers import Dense,Flatten,Input,GRU,Dropout,LSTM,Bidirectional
from keras import Sequential
from sklearn.ensemble import RandomForestClassifier
from keras.backend import clear_session
```

```python
clear_session()
model=Sequential()
model.add(Input(shape=(178,1)))
model.add(Dense(32,activation="tanh"))
#model.add(Dropout(0.3))
model.add(Dense(128,activation="tanh"))
model.add(Bidirectional(LSTM(128,activation="tanh")))
#model.add(Bidirectional(LSTM(128,activation="tanh")))
model.add(Dropout(0.2))
#model.add(BatchNormalization())
#model.add(BatchNormalization())
#model.add(Flatten())
model.add(Dense(64,activation="tanh"))
model.add(Dropout(0.2))
#model.add(BatchNormalization())
#model.add(BatchNormalization())
#model.add(Dense(128,activation="tanh"))
model.add(Flatten())
model.add(Dense(128,activation="tanh"))
#model.add(Dropout(0.2))
model.add(Dense(2,activation="softmax"))
model.compile(optimizer="Adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
model.summary()
```
